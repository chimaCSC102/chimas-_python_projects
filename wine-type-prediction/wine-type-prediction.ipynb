```python
#deep learning models are suitable for large datasets
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Double check for null values in `red`
# pd.isnull(red)


white_wine = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv", sep =';')
red_wine = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv", sep =';')
```

```python
#bar plot showing balance/imbalance
#white wine
bar_plot = white_wine #pd.read_csv('iris.csv').dropna(axis = 1)

species_counts = bar_plot["quality"].value_counts()
temp_df = pd.DataFrame({
    "quality" : species_counts.index,
    "Counts" : species_counts.values
})

plt.figure(figsize = (18,8))


#it's plotting the counts against the possible diseases
sns.barplot(x = "quality", y = "Counts", data = temp_df)
plt.xticks(rotation= 0)
plt.show()

#therefore the labels are unbalanced
```

```python
#bar plot showing balance/imbalance
#red wine
bar_plot = red_wine #pd.read_csv('iris.csv').dropna(axis = 1)

species_counts = bar_plot["quality"].value_counts()
temp_df = pd.DataFrame({
    "quality" : species_counts.index,
    "Counts" : species_counts.values
})

plt.figure(figsize = (18,8))


#it's plotting the counts against the possible diseases
sns.barplot(x = "quality", y = "Counts", data = temp_df)
plt.xticks(rotation= 0)
plt.show()

#therefore the labels are unbalanced
```

```python
#to add extra columns to the existing csv
red_wine['type'] = 1
white_wine['type'] = 0

#this part joins the csv's together
wines = pd.concat([red_wine,white_wine], ignore_index=True)
wines
```

```python
from sklearn.model_selection import train_test_split
features = wines.iloc[:, 0:11]

#labels are flattened when dealing with neural networks
label = np.ravel(wines.type)

X_train, X_test, y_train, y_test = train_test_split(features, label, test_size = 0.2, random_state=45)
```

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()


#features being fed in

#input
model.add(Dense(12, activation = 'relu', input_shape=(11,1)))


#hidden
model.add(Dense(9, activation='relu'))

#output
model.add(Dense(1, activation='sigmoid'))

model.output_shape
model.summary()
model.get_config()


model.get_weights()
model.compile(loss='binary_crossentropy', optimizer='adam',metrics=['accuracy'])
```
