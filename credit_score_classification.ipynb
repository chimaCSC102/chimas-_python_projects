```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
pio.templates.default = "plotly_white"

df = pd.read_csv("train.csv")

fig = px.box(df, 
             x="Credit_Score", 
             y="Num_of_Loan", 
             color="Credit_Score",
             title="Credit Scores Based on num of bank accounts", 
             color_discrete_map={'Poor':'red',
                                 'Standard':'yellow',
                                 'Good':'green'})
fig.update_traces(quartilemethod="exclusive")
fig.show()

df = df.drop(['Type_of_Loan','Credit_Mix','Num_Credit_Inquiries','Payment_Behaviour', 'Payment_of_Min_Amount', 'Total_EMI_per_month','Amount_invested_monthly', 'Credit_Utilization_Ratio','Changed_Credit_Limit','ID', 'Customer_ID', 'Name','Month','Age','SSN','Occupation'], axis=1)


'''
df['Credit_Score'] = df['Credit_Score'].map({'Standard': 1, 
                               'Good': 2, 
                               'Bad': 0})

df['Credit_Mix'] = df['Credit_Mix'].map({'Standard': 1, 
                               'Good': 2, 
                               'Bad': 0})
                               '''
df = df.fillna(0)
print(df.info())
print(df.isnull().sum()) #to check for null values
#df = df.drop(54617)
print(df["Credit_Score"].value_counts()) #to count the amount of good, bad, and standard
print(df.head(5))
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, f_regression

#Name Occupation Credit_Mix Payment_of_Min_Amount Payment_Behaviour Credit_Score  (non numeric data)
##conversion
le = LabelEncoder()

df['Credit_Score'] = le.fit_transform(df['Credit_Score'])

print(df)
##split
train_dataset, test_dataset = train_test_split(df, test_size=0.3, random_state=42)

train_labels = train_dataset.pop('Credit_Score')
test_labels = test_dataset.pop('Credit_Score')


#using a polynomial regression model

from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(train_dataset, train_labels)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

##model prediction

from sklearn.metrics import accuracy_score
y_pred = model.predict(test_dataset)
print(y_pred)
accuracy = accuracy_score(test_labels, y_pred)
print(f"Accuracy: {accuracy}")



```
