```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.linear_model import LinearRegression, LogisticRegression
```

```python
data = pd.read_csv('framingham.csv')
```

```python
data.dropna(axis = 0, inplace = True) #to remove null values
data.describe()
# data.head(5)
```

```python
#bar plot showing balance/imbalance of the label CHD

bar_plot = pd.read_csv('framingham.csv').dropna(axis = 1)

CHD_counts = bar_plot["TenYearCHD"].value_counts()
temp_df = pd.DataFrame({
    "CHD" : CHD_counts.index,
    "Counts" : CHD_counts.values
})

plt.figure(figsize = (18,8))


#it's plotting the counts against the possible diseases
sns.barplot(x = "CHD", y = "Counts", data = temp_df)
plt.xticks(rotation= 180)
plt.show()

#there are 41 unique diseases and they occur 120 times, thus there are 4920(120 * 41) entries
#balanced


#scatter plot- to check if a linear relationship exists between various features and it's label
# plt.scatter(x,y)
# plt.xlabel(x)
# plt.ylabel(y)
# plt.title(Scatter plot)
# plot.show()
#x.corr() to get correlation coefficient
```

```python
features = data.iloc[:,:-1]
label = data.iloc[:,-1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, label, test_size=0.2, random_state=24)
X_train.shape,y_train.shape, X_test.shape, y_test.shape
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
prediction = model.predict(X_test)
pd.DataFrame(prediction, y_test)

accuracy = accuracy_score(y_test,prediction)*100
print("Accuracy: ",accuracy)

#task accuracy was 0.8490230905861457
```
