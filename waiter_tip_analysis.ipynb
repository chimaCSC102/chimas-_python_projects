```python
import pandas as pd
import numpy as np

data = pd.read_csv("text.txt")
print(data.head(5))
print(data.info())
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split

le = LabelEncoder()
data['sex'] = le.fit_transform(data['sex'])
data['smoker'] = le.fit_transform(data['smoker'])
data['day'] = le.fit_transform(data['day'])
data['time'] = le.fit_transform(data['time'])

print(data)
```

```python
features = data[['total_bill','sex','smoker','day','time','size']]
print(features)

label = data[['tip']]
print(label)
```

```python
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
pio.templates.default = "plotly_white"

fig = px.box(data, 
             x="total_bill",  
             color="tip", 
             title="tip Based on total_bill", 
             color_discrete_map={'Poor':'red',
                                 'Standard':'yellow',
                                 'Good':'green'})
fig.show()


fig = px.box(data, 
             x="sex",  
             color="tip", 
             title="tip Based on sex", 
             color_discrete_map={'Poor':'red',
                                 'Standard':'yellow',
                                 'Good':'green'})
fig.show()



fig = px.box(data, 
             x="smoker",  
             color="tip", 
             title="tip Based on smoker", 
             color_discrete_map={'Poor':'red',
                                 'Standard':'yellow',
                                 'Good':'green'})
fig.show()


fig = px.box(data, 
             x="day",  
             color="tip", 
             title="tip Based on day", 
             color_discrete_map={'Poor':'red',
                                 'Standard':'yellow',
                                 'Good':'green'})
fig.show()


fig = px.box(data, 
             x="time",  
             color="tip", 
             title="tip Based on time", 
             color_discrete_map={'Poor':'red',
                                 'Standard':'yellow',
                                 'Good':'green'})
fig.show()


fig = px.box(data, 
             x="size",  
             color="tip", 
             title="tip Based on size", 
             color_discrete_map={'Poor':'red',
                                 'Standard':'yellow',
                                 'Good':'green'})
fig.show()
```

```python
model = LinearRegression()
model.fit(features, label)
```

```python
input = [[24.50, 1, 0, 0, 1, 4]]
prediction = model.predict(input)
prediction = float(prediction)
print("The predicted tip is: ", prediction)
```
